### Highly divisible triangular number
The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 1: 1
 
 3: 1,3
 
 6: 1,2,3,6

10: 1,2,5,10

15: 1,3,5,15

21: 1,3,7,21

28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

###### Solution

```python
def smallest_triangle(min_divisors):
    """
    Returns the smallest triangle number with more than a given number of divisors
    
    Input: int
    
    Returns: int
    """
    triangle_number = 1
    addition = 1
    divisors = 0
    
    while divisors <= min_divisors:
        divisors = 0
        addition += 1
        triangle_number += addition
        div = 1
        while div*div < triangle_number:
            if triangle_number%div == 0:
                divisors += 2
            div += 1
        if div*div == triangle_number:
            divisors += 1
    
    return triangle_number
```

```python
smallest_triangle(5)
```

```python
smallest_triangle(500)
```
